**Multiline**

    Physical lines of code end with physical newline character
    Logical lines of code end with a logical newline token
    
    Sometimes, physical newlines are ignored in order to combine multiple physical lines into a single logical line of code terminated by a logical NEWLINE token. Which allow us to write code over multiple line.
    
    It's important to make code readable which will be helpful to us when we will be reading the code later.
    
    [1, #item 1
     2, #item 2
     3 #item 3
    ]
    
    def my_func(a, 
                b, #comment
                c):
        print(a, b, c)
        
    def my_func(10, #comment
                20, 30)
                
    At the end this comments will be stripped out and entire multilines will be converted into single logical newline.
